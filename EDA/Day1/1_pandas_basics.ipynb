## Difference Between `loc` and `iloc`

### `iloc`
- Stands for **integer-location based indexing**.
- It is used for selecting rows and columns **by their numerical positions** (integer indices).
- **Example**: 
  ```python
  df.iloc[1:4, 0:2]
  
### `loc`

- Stands for **label-location based indexing**.
- It is used for selecting rows and columns **by their labels** (e.g., column names or row labels).
- **Example**: 
  ```python
  df.loc[1:3, ['A', 'B']]


````python
df.loc[1:50:3,['RelationshipSatisfaction', 'JobRole', 'StockOptionLevel']]
# start, stop,stepper
````

````python
df.iloc[1:4,0:10]
````

### Creating a separate, small dataset

````python
df_new = df.loc[:100, ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department']]
df_new
````

````python
#will automatically create a column if it doesn't exist and add it at the end

df_new["new_col"] = df.loc[:100, 'JobRole']

df_new
````

````python
#adding a column at a specific location

df_new.insert(loc= 1, column = "new_col2", value = df["Department"])

df_new
````

````python
#to drop columns, you need to set axis = 1. by default it is set to 0, meaning rows

df_new=df_new.drop('new_col',axis=1)
df_new
````

````python
df_new.drop('new_col2',axis=1,inplace=True)
df_new
````
